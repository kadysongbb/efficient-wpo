### Wholesale Price 

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
sns.set_style("whitegrid")
%matplotlib inline
plt.rcParams["figure.figsize"] = (6,5)

ax = plt.subplot(111)
plt.ylabel('Wholesale Price', fontsize=22)
plt.xlabel('Hour', fontsize=22)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.xlim(1,24)
plt.ylim(0,10)
plt.locator_params(axis="x", nbins=5)

x = np.arange(1,25,1)
wholesale_price = np.asarray([2,2,2,2,2,2,2,2,2,2,3,3,3,3,3,3,6,6,6,4,3,3,2,2])
plt.plot(x, wholesale_price, marker = '*')
```
